**Objective :** This code needs to resize all the images "images/" into a specified format th

I implemented two methods : 
 - The first one is resizing the images into a square no matter what is the input file. There is no loss of information but if the input image is a rectangle, then the output file would look a bit compressed.
 - The second one is taking a square out of any rectangle and resizing it. There is a loss of information if the input picture is a rectangle. 

This is the part where we import everything that will be used later in the code.

```python
from PIL import Image
import os
```

Parameters

```python
input_dir          = "images/"
firstmethod_dir    = "output_firstmethod_images/"
secondmethod_dir   = "output_secondmethod_images/"
format             = (256, 256)
```

Code to create the folders if they don't exist

```python
# Ensure output directory exists
if not os.path.exists(firstmethod_dir):
    os.makedirs(firstmethod_dir)
if not os.path.exists(secondmethod_dir):
    os.makedirs(secondmethod_dir)
```

### Code for the first method

```python
i = 1
# Resize images
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
        # Open image
        img = Image.open(os.path.join(input_dir, filename))
        
        # Resize image 
        img_resized = img.resize(format, Image.LANCZOS)
        
        # Save resized image
        output_path = os.path.join(firstmethod_dir, i.__str__() + ".png")
        img_resized.save(output_path)
        i = i+1
        print(f"Resized {filename} to {format} format.")
```

### Code for the second method

```python
i = 1
for filename in os.listdir(input_dir):
    if filename.endswith(".jpg") or filename.endswith(".png") or filename.endswith(".jpeg"):
        # Open image
        img = Image.open(os.path.join(input_dir, filename))
        
        # Find the square region to extract
        width, height = img.size
        min_dim = min(width, height)
        left = (width - min_dim) // 2
        top = (height - min_dim) // 2
        right = left + min_dim
        bottom = top + min_dim
        
        # Extract square region
        img_square = img.crop((left, top, right, bottom))
        
        # Resize image
        img_resized = img_square.resize(format, Image.LANCZOS)
        
        # Save resized image
        output_path = os.path.join(secondmethod_dir, i.__str__() + ".png")
        img_resized.save(output_path)
        i = i+1
        print(f"Resized {filename} to {format} format.")
```
